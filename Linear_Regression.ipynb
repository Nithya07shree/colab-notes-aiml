```python
import numpy as np
```

```python
class linearRegression:
  def __init__(self, learningRate=0.01, iterations=1000):
    self.learningRate = learningRate
    self.iterations = iterations
    self.weights = None
    self.bias = None
  def fit(self, X,y):
    n_samples, n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias=0

    for _ in range(self.iterations):
      # prediction
      y_pred = np.dot(X, self.weights) + self.bias
      # calculate gradient
      dw = (1/n_samples) * np.dot(X.T, (y_pred - y))
      db = (1/n_samples) * np.sum(y_pred - y)
      #update weights
      self.weights -= self.learningRate*dw
      self.bias -= self.learningRate*db
  def predict(self, X):
    y_pred = np.dot(X, self.weights) + self.bias
    return y_pred
```

```python
x=np.linspace(0,1,100)
noise = np.random.randn(100)
y=3*x+5+noise
```

```python
print(x[:5])
```

```python
lr=linearRegression(learningRate=0.0001, iterations=1000)
```

```python
X = x.reshape(-1, 1)
```

```python
lr.fit(X,y)
```

```python
y_pred = lr.predict(X)
```

```python
print(f"Final Weights: {lr.weights}")
```

```python
print(f"Final Bias: {lr.bias}")
```
