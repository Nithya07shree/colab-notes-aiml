Create tensors

```python
import torch
```

```python
tensor1 = torch.tensor([1,2,3])
tensor2 = torch.randn(2,3)
tensor3 = torch.randn_like(tensor1, dtype=torch.float)
```

```python
print(f"tensor1: {tensor1}")
print(f"tensor2: {tensor2}")
print(f"tensor3:  {tensor3}")
```

parameters of tensors

```python
print(f"shape: {tensor3.shape}")
print(f"dtype: {tensor3.dtype}")
print(f"device: {tensor3.device}")
```

Autograd- building the graph

```python
# requires_grad = True
a=torch.tensor(2.0, requires_grad=True)
b=torch.tensor(1.0, requires_grad=True)
x=torch.tensor(3.0, requires_grad=True)

y=a+b
z=x*y
```

```python
#how do u know the graph exists
# .grad_fn
print(y.grad_fn)
print(z.grad_fn)
print(a.grad_fn)
```

operations - verbs of pytorch

```python
# * vs @
t1 = torch.tensor([[1.0,2.0],[3.0,4.0]])
t2 = torch.tensor([[5.0,6.0],[7.0,8.0]])

print(t1*t2) # element wise multiplication
print(t1@t2) # matrix multiplication
```

```python
# reduction
# sum, mean, max
# in 2d tensors, dim = 0 operates vertically(cols), dim = 1 operates horizontally(rows)
# t1.mean(dim=0)
print(f"mean: {t1.mean()}")
print(f"sum: {t1.sum()}")
print(f"max: {t1.max()}")
```

Selecting your data

```python
#  basic indexing
x = torch.arange(12).reshape(3,4)
print(x)
print(x[:,2]) # get third column
print(x[1,3]) # get 1st row 3rd column
```

```python
# argmax - find index of highest value
a = torch.tensor([[1,2,4,2,1],[12,3,43,2,5]])
print(a.argmax(dim=0))
print(a.argmax(dim=1))
```

```python
# torch.gather()
data = torch.tensor([
    [1,2,3,4],
    [2,3,4,5],
    [3,4,5,6]
])
selection = torch.tensor([[0],[1],[2]])
print(torch.gather(data,dim=1,index=selection))
```
